The files used in this report are generated by the script we wrote. The code is on /cmd/test/main.go. To generate the .txt file you need to uncomment file writing and run `make run-test`. 

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Read data from writeTimes.txt and readTimes.txt files
with open('writeTimes.txt', 'r') as write_file:
    write_times = [float(line.strip()) for line in write_file.readlines()]

with open('readTimes.txt', 'r') as read_file:
    read_times = [float(line.strip()) for line in read_file.readlines()]

# Sample data (replace with your actual data)
read_times = np.array(read_times)  # NumPy array of read times in nanoseconds
write_times = np.array(write_times)  # NumPy array of write times in nanoseconds
```

## Write data

```python
plt.figure(figsize=(8, 6))
sns.histplot(write_times, bins=20, kde=True)
plt.xlabel('Time (nanoseconds)')
plt.ylabel('Frequency')
plt.title('Histogram of Read Times')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=write_times)
plt.ylabel('Time (nanoseconds)')
plt.title('Distribution of Read Operation Durations')
plt.show()
```

## Read data

```python
plt.figure(figsize=(8, 6))
sns.histplot(read_times, bins=20, kde=True)
plt.xlabel('Time (nanoseconds)')
plt.ylabel('Frequency')
plt.title('Histogram of Read Times')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=read_times)
plt.ylabel('Time (nanoseconds)')
plt.title('Distribution of Read Operation Durations')
plt.show()
```
